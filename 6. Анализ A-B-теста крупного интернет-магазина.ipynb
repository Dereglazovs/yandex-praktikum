```python
hypot['ICE'] = (hypot.impact * hypot.confidence) / hypot.efforts
hypot_ice = hypot[['hypothesis', 'ICE']].sort_values(by = 'ICE', ascending = False).round(2).reset_index().drop(columns = 'index')
hypot_ice
```

<a name = "2.2."></a><font size = "3"><b>2.2 Применение фреймворка RICE для приоритизации гипотез и сортировка их по убыванию приоритета</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

```python
hypot['RICE'] = (hypot.reach * hypot.impact * hypot.confidence) / hypot.efforts
hypot_rice = hypot[['hypothesis', 'RICE']].sort_values(by = 'RICE', ascending = False).round(2).reset_index().drop(columns = 'index')
hypot_rice
```

<a name = "3.1."></a><font size = "3"><b>3.1 Построение графика кумулятивной выручки по группам</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

```python
# получим агрегированные кумулятивные данные о заказах 
orderAggregated = order.apply(
    lambda x: order[np.logical_and(order['date'] <= x['date'], order['group'] == x['group'])]
    .agg({'date' : 'max',
          'group' : 'max',
          'transaction_id' : pd.Series.nunique,
          'visitor_id' : pd.Series.nunique,
          'revenue' : 'sum'
         }), axis = 1
).sort_values(by = ['date', 'group'])

# получим агрегированные кумулятивные данные о пользователях
visitAggregated = visit.apply(
    lambda x: visit[np.logical_and(visit['date'] <= x['date'], visit['group'] == x['group'])]
    .agg({'date' : 'max',
          'group' : 'max',
          'visitors' : 'sum'
         }), axis = 1
).sort_values(by = ['date', 'group'])

# объединим кумулятивные данные в одной таблице
cumData = orderAggregated.merge(
    visitAggregated, left_on = ['date', 'group'], right_on = ['date', 'group']
)

# переименнуем название столбцов
cumData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

cumData.head()
```

```python
# датафрейм с кумулятивным количеством заказов и выручкой А и B групп
cum_rev_A = cumData[cumData['group'] == 'A'][['date', 'revenue', 'orders']]
cum_rev_B = cumData[cumData['group'] == 'B'][['date', 'revenue', 'orders']]
display(cum_rev_A.head())
display(cum_rev_A.head())
```

```python
plt.figure(figsize = (15, 7))
plt.plot(cum_rev_A['date'], cum_rev_A['revenue'], label = 'A')
plt.plot(cum_rev_B['date'], cum_rev_B['revenue'], label = 'B')

plt.minorticks_on()
plt.grid(which='minor', linestyle=':')
plt.legend()
plt.title('График кумулятивной выручки по группам', fontsize = 15)
plt.xlabel('Дата')
plt.ylabel('Стоимость заказа')
plt.show()
```

**Вывод:** 

Практически на протяжении всего теста группы идут параллельно др. др. В группа B заказов больше и в период с 18 августа по 19 августа был большой скачок заказов. Аномально большие заказы влияют на результаты.

<a name = "3.2."></a><font size = "3"><b>3.2 Построение графика кумулятивного среднего чека по группам</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

```python
plt.figure(figsize = (15, 7))
plt.plot(cum_rev_A['date'], cum_rev_A['revenue'] / cum_rev_A['orders'], label = 'A')
plt.plot(cum_rev_B['date'], cum_rev_B['revenue'] / cum_rev_B['orders'], label = 'B')

plt.minorticks_on()
plt.grid(which='minor', linestyle=':')
plt.legend()
plt.title('График кумулятивного среднего чека', fontsize = 15)
plt.xlabel('Дата')
plt.ylabel('Средняя стоимость заказа')
plt.show()
```

**Вывод:**

На графике линии несколько раз пересекаются и видны резкие скачки заказов. Необходимо посмотреть на выбросы (дорогие покупки).

<a name = "3.3."></a><font size = "3"><b>3.3 Построение графика относительного изменения кумулятивного среднего чека группы B к группе A</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

```python
merged_cum_rev = cum_rev_A.merge(
    cum_rev_B, left_on = 'date', right_on = 'date', how = 'left', suffixes = ['A', 'B']
)
plt.figure(figsize = (15, 7))
plt.plot(merged_cum_rev['date'], (merged_cum_rev['revenueB'] / merged_cum_rev['ordersB']) / (merged_cum_rev['revenueA'] / merged_cum_rev['ordersA']) - 1)

plt.minorticks_on()
plt.grid(which='minor', linestyle=':')
plt.title('График относительного изменения кумулятивного среднего чека группы B к группе A', fontsize = 15)
plt.xlabel('Дата')
plt.ylabel('Средняя стоимость заказа')
plt.axhline(y = 0, color = 'red', linestyle = '--')
plt.show()
```

**Вывод:**

Как и отмечалось ранее имеются аномальные выбросы. В места резкого роста совершались покупки на крупные суммы.

<a name = "3.4."></a><font size = "3"><b>3.4 Построение графика кумулятивной конверсии по группам</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

```python
cumData['conversion'] = cumData['orders'] / cumData['visitors']

cum_data_A = cumData[cumData['group'] == 'A']
cum_data_B = cumData[cumData['group'] == 'B']

plt.figure(figsize = (15, 7))
plt.plot(cum_data_A['date'], cum_data_A['conversion'], label = 'A')
plt.plot(cum_data_B['date'], cum_data_B['conversion'], label = 'B')

plt.minorticks_on()
plt.grid(which='minor', linestyle=':')
plt.title('График кумулятивной конверсии по группам', fontsize = 15)
plt.xlabel('Дата')
plt.ylabel('Количество возвратов')
plt.legend()
plt.show()
```

**Вывод:**

До 06 августа группа А имела большую конверсию, чем группа B. Но после это числа вторая группа поднялась до отметки 0,0365 и далее постепенно снижалась. При этом конверсия группы B стала больше группы А до конца теста.

<a name = "3.8."></a><font size = "3"><b>3.8 Подсчет статистической значимости различий в конверсии и среднем чеке между группами по «сырым» данным</b></font><br/>
<font size = "2">[содержание](#0.0)</font><br/>

**Сформируем гипотезы:**<br/>
<i>Нулева гипотеза:</i> Cтатистической значимости различий в конверсии между группами нет<br/>
<i>Альтернативная гипотеза:</i> Cтатистическая значимость различий в конверсии между группами есть<br/>

<i>Нулева гипотеза:</i> Cтатистической значимости различий в среднем чеке между группами нет<br/>
<i>Альтернативная гипотеза:</i> Cтатистическая значимость различий в среднем чеке между группами есть<br/>
alpha = .05

```python
# количество пользователей до выбранной даты включительно в группе A
visitorsADaily = visit[visit['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA'] # количество пользователей

# накопленное количество посетителей по дням для группы А
visit_cum_A = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date' : 'max', 'visitorsPerDateA' : 'sum'}
    ),
    axis=1,
)
visit_cum_A.columns = ['date', 'visit_cum_A']

# количество пользователей до выбранной даты включительно в группе B
visitorsBDaily = visit[visit['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB'] # количество пользователей

# накопленное количество посетителей по дням для группы В
visit_cum_B = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date' : 'max', 'visitorsPerDateB' : 'sum'}
    ),
    axis=1,
)
visit_cum_B.columns = ['date', 'visit_cum_B']

# количество заказов по дням длягруппы А
ordersADaily = (
    order[order['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id' : pd.Series.nunique, 'revenue' : 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA'] # количество заказов, суммарная выручка

# суммарное число заказов для группы А
order_cum_A = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date' : 'max', 'ordersPerDateA' : 'sum', 'revenuePerDateA' : 'sum'}
    ),
    axis=1,
).sort_values(by = ['date'])

order_cum_A.columns = [
    'date',
    'order_cum_A',
    'rev_cum_A', 
] # суммарное число заказов, суммарная выручка

# количество заказов по дням длягруппы B
ordersBDaily = (
    order[order['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id' : pd.Series.nunique, 'revenue' : 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB'] # количество заказов, суммарная выручка

# суммарное число заказов для группы B
order_cum_B = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date' : 'max', 'ordersPerDateB' : 'sum', 'revenuePerDateB' : 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])

order_cum_B.columns = [
    'date',
    'order_cum_B',
    'rev_cum_B',
] # суммарное число заказов, суммарная выручка

# объединение данных в единую таблицу
dt = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(order_cum_A, left_on='date', right_on='date', how='left')
    .merge(order_cum_B, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visit_cum_A, left_on='date', right_on='date', how='left')
    .merge(visit_cum_B, left_on='date', right_on='date', how='left')
)
dt.head()
```

```python
# подсчет конверсии
# посчитаем, для пользователей которые совершали хотя бы один заказ, количество совершенных заказов
order_user_A = (
    order[order['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
order_user_A.columns = ['user_id', 'orders']

order_user_B = (
    order[order['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
order_user_B.columns = ['user_id', 'orders']
```

```python
# Введем переменные sam_A и sam_B, в которых пользователям из разных групп будет соответствовать количество заказов
#Тем, кто ничего не заказал, будут соответствовать нули
# Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни
sam_A = pd.concat([order_user_A['orders'], pd.Series(0, index=np.arange(dt['visitorsPerDateA'].sum() - len(order_user_A['orders'])), name = 'orders')], axis = 0)
sam_B = pd.concat([order_user_B['orders'], pd.Series(0, index=np.arange(dt['visitorsPerDateB'].sum() - len(order_user_B['orders'])), name = 'orders')], axis = 0)
```

```python
display('p-value: {0:.3f}'.format(st.mannwhitneyu(sam_A, sam_B)[1]))
display('{0:.3f}'.format(sam_B.mean() / sam_A.mean() - 1))
```

**Вывод:**

- по "сырым" данным можно сказать, что между группами существуют различия в конверсии;
- p-value получилось 0.017 меньше 0.05, а значит отвергаем нулевую гипотезу: Cтатистической значимости различий в конверсии между группами нет. Относительный прирост группы B равен 13.8%.

```python
# подсчет среднего чека
display('p-value: {0:.3f}'.format(st.mannwhitneyu(order[order['group']=='A']['revenue'], order[order['group']=='B']['revenue'])[1]))
display('{0:.3f}'.format(order[order['group']=='B']['revenue'].mean()/order[order['group']=='A']['revenue'].mean()-1)) 
```

**Вывод:**

- по "сырым" данным можно сказать, что между группами различий в среднем чеке нет;
- p-value получилось 0.729 больше 0.05, а значит отвергать нулевую гипотезу незачем: Cтатистической значимости различий в среднем чеке между группами нет. Относительный прирост группы B равен 25.9%.

<a name = "3.9."></a><font size = "3"><b>3.9 Подсчет статистической значимости различий в конверсии и среднем чеке заказа между группами по «очищенным» данным</b></font><br/>

Ранее были выявлены аномальные значения:<br/>
- количество заказов больше 2;
- сумма заказа больше 28 тыс.

```python
order.head()
```

```python
user_order = pd.concat(
    [
        order_user_A[order_user_A['orders'] > 2]['user_id'],
        order_user_B[order_user_B['orders'] > 2]['user_id'], 
    ],
    axis=0,
)
user_rev_order = order[order['revenue'] > 28000]['visitor_id'] 

abnormal_user = (
    pd.concat([user_order, user_rev_order], axis=0)
    .drop_duplicates()
    .sort_values()
)

display(abnormal_user.head(5))
display('Всего аномальных пользователей:', abnormal_user.shape[0])
```

<div class="alert alert-block alert-warning"><b>Комментарий ревьюера</b>
    
<b>Совет:</b><br>
У тебя опечатка в строчке - **order_user_B[order_user_B['orders'] > 2]['orders']** <br>
Здесь надо выбрать id покупателей, т.е. **user_id**    <br>
И в строчке   **user_rev_order = order[order['revenue'] > 28000]['transaction_id']** <br>
Здесь надо выбрать id покупателей, т.е. **visitor_id** 
</div>

```python
# подсчет конверсии
sam_A_filter = pd.concat(
    [
        order_user_A[
            np.logical_not(order_user_A['user_id'].isin(abnormal_user))
        ]['orders'],
        pd.Series(
            0,
            index = np.arange(
                dt['visitorsPerDateA'].sum() - len(order_user_A['orders'])
            ),
            name = 'orders',
        ),
    ],
    axis=0,
)

sam_B_filter = pd.concat(
    [
        order_user_B[
            np.logical_not(order_user_B['user_id'].isin(abnormal_user))
        ]['orders'],
        pd.Series(
            0,
            index = np.arange(
                dt['visitorsPerDateB'].sum() - len(order_user_B['orders'])
            ),
            name = 'orders',
        ),
    ],
    axis=0,
) 
```

```python
display('p-value: {0:.3f}'.format(st.mannwhitneyu(sam_A_filter, sam_B_filter)[1]))
display('{0:.3f}'.format(sam_B_filter.mean() / sam_A_filter.mean() - 1))
```

**Вывод:**

- по "очищенным" данным можно сказать, что между группами существуют различия в конверсии;
- p-value получилось 0.013 меньше 0.05, а значит отвергаем нулевую гипотезу: Cтатистической значимости различий в конверсии между группами нет. Относительный прирост группы B равен 20.5%.

```python
# подсчет среднего чека
display(
    'P-value: {0:.3f}'.format(
        st.mannwhitneyu(
            order[
                np.logical_and(
                    order['group'] == 'A',
                    np.logical_not(order['visitor_id'].isin(abnormal_user)),
                )
            ]['revenue'],
            order[
                np.logical_and(
                    order['group'] == 'B',
                    np.logical_not(order['visitor_id'].isin(abnormal_user)),
                )
            ]['revenue'],
        )[1]
    )
)

display(
    "{0:.3f}".format(
        order[
            np.logical_and(
                order['group'] == 'B',
                np.logical_not(order['visitor_id'].isin(abnormal_user)),
            )
        ]['revenue'].mean()
        / order[
            np.logical_and(
                order['group'] == 'A',
                np.logical_not(order['visitor_id'].isin(abnormal_user)),
            )
        ]['revenue'].mean()
        - 1
    )
) 
```

**Вывод:**

- по "очищенным" данным можно сказать, что между группами различий в среднем чеке нет;
- p-value получилось 0.738 больше 0.05, а значит отвергать нулевую гипотезу незачем: Cтатистической значимости различий в среднем чеке между группами нет. Относительный прирост группы B равен 22.3%.
- по "очищенным" значения поменялись, но не значительно и на результат это никак не повлияло.
