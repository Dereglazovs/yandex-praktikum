Согласно своду правил 54.13330.2016 Здания жилые многоквартирные площадь квартир должна быть не менее:

► Общей жилой комнаты в однокомнатной квартире — 14 м².

► Общей жилой комнаты в квартирах с числом комнат две и более — 16 м². 

► Спальни — 8 м² (на двух человек — 10 м²); кухни — 8 м².

► Кухонной зоны в кухне-столовой — 6 м².

► В однокомнатных квартирах допускается проектировать кухни или кухни-ниши площадью не менее 5 м².


Рекомендуемая площадь квартир:

* 1 - 28-38 м. кв.

* 2 - 44-53 м. кв.

* 3 - 56-65 м. кв.

* 4 - 70-77 м. кв.

* 5 - 84-96 м. кв.

* 6 - 103-109 м. кв.

Но фактически застройщик может проектировать площадь квартир по своему желанию.

```python
# посмотрим есть ли строки, где жилая пл. и пл. кухни больше общей пл., и где пл. кухни больше или равна жилой пл.
living_kitchen = df.query('(living_area + kitchen_area) > total_area')
living_kitchen
```

```python
# удалим некорректные данные
living_kitchen_drop = living_kitchen.index
df.drop(living_kitchen_drop, inplace=True)
df.info()
```

```python
kitchen = df.query('kitchen_area >= living_area')
kitchen[['kitchen_area', 'living_area', 'total_area']].head()
```

Заменим пустые строки медианным значением по отношению к количеству комнат в квартире.

```python
# заменим медианным значением площадь кухни
list_category = df[df['kitchen_area'].isna()]['rooms'].unique()
for o in list_category:
    value_for_filling = df.groupby('rooms')['kitchen_area'].median().get(key = o)
    df['kitchen_area'] = df.groupby('rooms')['kitchen_area'].apply(lambda x: x.fillna(x.dropna().median()))
```

```python
# заменим медианным значением жилую площадь
list_category = df[df['living_area'].isna()]['rooms'].unique()
for o in list_category:
    value_for_filling = df.groupby('rooms')['living_area'].median().get(key = o)
    df['living_area'] = df.groupby('rooms')['living_area'].apply(lambda x: x.fillna(x.dropna().median()))
df.info()
```

Попробуем восстановить данные с помощью столбца city_centers_nearest, т.е. чем дальше от центра будут квартиры, тем ниже будет высота потолков.

Возьмем медианное значение расстояние от центра города - поделим на группы все строки столбца city_centers_nearest.

```python
# найдем медиану и округлим
df.query('locality_name == "санкт-петербург"')['city_centers_nearest'].median()
```

```python
df['ceiling_height'].isna().value_counts() # True пропущенные значения
```

```python
# округлили до 12300
def distance(row):
    if row['city_centers_nearest'] < 3000:
        return 'центр до 3 км'
    if row['city_centers_nearest'] < 6000:
        return 'центр до 6 км'
    if row['city_centers_nearest'] <= 12300:
        return 'в пределах города'
    if row['city_centers_nearest'] > 12300:
        return 'область'
df = df.reset_index()
df['city_centers_category'] = df.apply(distance, axis = 1)
df.head()
```

```python
df.groupby('city_centers_category')['ceiling_height'].median().sort_values(ascending=False)
# видим, что чем дальше от центра тем высота потолков ниже
```

<a name = "2.3.8."></a>**2.3.8 Замена пустых ячеек в ceiling_height медианным значением по столбцу city_centers_category**<br/> 
<font size = "2">[содержание](#0.0)</font>

Высота потолков варьируется от 2.6 до 3.2 м., можно сказать, что заполнение близким значением пустых строк будет оптимальным.

```python
def index(category):
    median = df[df['city_centers_category'] == category]['ceiling_height'].median()
    return median

list_category = df[df['ceiling_height'].isna()]['city_centers_category'].unique()

for category in list_category:
    median = index(category)
    row = (df['city_centers_category'] == category) & (df['ceiling_height'].isna())
    df.loc[row, 'ceiling_height'] = df.loc[row, 'ceiling_height'].fillna(median)
df.info()
```
